```python
from mini_gpr import kernels
from mini_gpr.viz import sample_kernel

%config InlineBackend.figure_format = "svg"


sample_kernel(kernels.Linear(m=2, scale=1), n_samples=7, seed=42)
```

```python
k = kernels.Linear(m=2, scale=1) ** 2
sample_kernel(k, n_samples=4, seed=42)
```

```python
k = kernels.Constant(1)
sample_kernel(k, n_samples=7, seed=42)
```

```python
k = kernels.Periodic(period=3, lengthscale=1)
sample_kernel(k, n_samples=1, seed=42)
```

```python
k = kernels.Periodic(period=1, lengthscale=3)
sample_kernel(k, n_samples=1, seed=42)
```

```python
k = kernels.Periodic(period=1, lengthscale=3) + kernels.Linear(m=2, scale=1)
sample_kernel(k, n_samples=1, seed=42)
```

```python
k = kernels.Periodic(period=1, lengthscale=3) * kernels.Linear(m=3)
sample_kernel(k, n_samples=1, seed=42)
```
